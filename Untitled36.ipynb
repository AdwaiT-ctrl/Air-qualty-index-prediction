```python
import torch

def save_checkpoint(model, optimizer, epoch, train_acc, val_acc, best_val_acc, filepath):
    """Save training checkpoint"""
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_acc': train_acc,
        'val_acc': val_acc,
        'best_val_acc': best_val_acc,
    }
    torch.save(checkpoint, filepath)
    print(f"âœ“ Checkpoint saved: epoch {epoch+1}")

def load_checkpoint(filepath, model, optimizer):
    """Load checkpoint and resume"""
    if not os.path.exists(filepath):
        return 0, 0.0  # start_epoch, best_val_acc

    checkpoint = torch.load(filepath)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])

    print(f"âœ“ Resumed from epoch {checkpoint['epoch']+1}")
    return checkpoint['epoch'] + 1, checkpoint.get('best_val_acc', 0.0)
```

```python
# Configuration
checkpoint_dir = '/content/drive/MyDrive/food_classification_checkpoints'
TOTAL_EPOCHS = 10
LEARNING_RATE = 0.0001

# Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Try to resume from checkpoint
checkpoint_path = f'{checkpoint_dir}/checkpoint_latest.pth'
start_epoch, best_val_acc = load_checkpoint(checkpoint_path, model, optimizer)

# Training loop
for epoch in range(start_epoch, TOTAL_EPOCHS):
    print(f"\nEpoch [{epoch+1}/{TOTAL_EPOCHS}]")

    # === TRAINING PHASE ===
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        if (i+1) % 150 == 0:
            print(f"  Batch {i+1}/{len(train_loader)} - Loss: {loss.item():.4f}")

    train_acc = 100. * correct / total

    # === VALIDATION PHASE ===
    model.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    val_acc = 100. * val_correct / val_total

    # Update best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), f'{checkpoint_dir}/best_model.pth')
        print(f"ðŸŽ‰ New best: {best_val_acc:.2f}%")

    print(f"Train: {train_acc:.2f}% | Val: {val_acc:.2f}% | Best: {best_val_acc:.2f}%")

    # === SAVE CHECKPOINT ===
    save_checkpoint(model, optimizer, epoch, train_acc, val_acc,
                   best_val_acc, checkpoint_path)
```

```python
# Define Transforms
# We use standard ImageNet normalization because we are using a pre-trained model
stats = ((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))

train_transform = transforms.Compose([
    transforms.Resize((128, 128)),      # Resize for speed
    transforms.RandomHorizontalFlip(),  # Augmentation: Flip
    transforms.RandomRotation(10),      # Augmentation: Rotate
    transforms.ColorJitter(brightness=0.1, contrast=0.1), # Augmentation: Lighting
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])

test_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])

# Download and Load Data
print("Downloading Food-101 Dataset")
train_dataset = torchvision.datasets.Food101(root='./data', split='train',
                                             download=True, transform=train_transform)
test_dataset = torchvision.datasets.Food101(root='./data', split='test',
                                            download=True, transform=test_transform)

# Create DataLoaders
# Batch size 64 strikes a balance between speed and memory on T4 GPU
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True,
                          num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False,
                         num_workers=2, pin_memory=True)

print(f"Training Samples: {len(train_dataset)}")
print(f"Test Samples: {len(test_dataset)}")
print(f"Classes: {len(train_dataset.classes)}")
```

```python
def get_transfer_model(num_classes=101):
    # 1. Load pre-trained ResNet18
    # 'DEFAULT' loads the best available weights (ImageNet)
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

    # 2. Freeze all parameters (Disable training for feature layers)
    for param in model.parameters():
        param.requires_grad = False

    # 3. Replace the final fully connected layer (fc)
    # The original fc layer has 512 input features.
    # We create a new layer with 101 outputs (for our food classes).
    # Since we created this layer new, requires_grad is True by default.
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, num_classes)

    return model.to(device)

# Instantiate the model
model = get_transfer_model(num_classes=101)
print("Model created: ResNet18 with frozen backbone and new 101-class head.")
```

```python
def train_model(model, train_loader, test_loader, epochs=5, lr=0.001):

    optimizer = optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()), lr=lr
    )
    criterion = nn.CrossEntropyLoss()

    checkpoint_path = "/content/drive/MyDrive/food101_checkpoint.pth"

    best_val_loss = float('inf')

    print(f"Starting training for {epochs} epochs...")

    for epoch in range(epochs):

        model.train()
        running_loss = 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # ----- VALIDATION -----
        model.eval()
        val_loss = 0.0

        with torch.no_grad():
            for inputs, labels in test_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}] | "
              f"Train Loss: {running_loss:.4f} | "
              f"Val Loss: {val_loss:.4f}")

        # ----- SAVE CHECKPOINT EVERY EPOCH -----
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_loss': val_loss,
        }, checkpoint_path)

        # ----- SAVE BEST MODEL ONLY -----
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(),
                       "/content/drive/MyDrive/food101_best_model.pth")
            print("Best model saved!")
```

```python
def visualize_predictions(model, loader):
    model.eval()
    images, labels = next(iter(loader))
    images = images.to(device)

    # Get predictions
    outputs = model(images)
    _, preds = torch.max(outputs, 1)

    # Move back to CPU for plotting
    images = images.cpu()
    preds = preds.cpu()

    # Plot first 5 images
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))
    classes = loader.dataset.classes

    for i in range(5):
        # Un-normalize image for display
        img = images[i].permute(1, 2, 0) * torch.tensor(stats[1]).view(1, 1, 3) + torch.tensor(stats[0]).view(1, 1, 3)
        img = torch.clamp(img, 0, 1)

        axes[i].imshow(img)
        title_color = 'green' if preds[i] == labels[i] else 'red'
        axes[i].set_title(f"Pred: {classes[preds[i]]}\nReal: {classes[labels[i]]}", color=title_color, fontsize=10)
        axes[i].axis('off')
    plt.show()

visualize_predictions(model, test_loader)
```

```python
import requests

def check_image_url(url):
    try:
        response = requests.head(url, allow_redirects=True, timeout=5) # Use HEAD request for efficiency
        if response.status_code == 200:
            content_type = response.headers.get('Content-Type', '')
            if 'image' in content_type:
                print(f"URL: {url}\nStatus: OK ({response.status_code})\nContent-Type: {content_type}\nResult: Valid image URL")
                return True
            else:
                print(f"URL: {url}\nStatus: OK ({response.status_code})\nContent-Type: {content_type}\nResult: Not a direct image link (Content-Type is not image)")
                return False
        else:
            print(f"URL: {url}\nStatus: Error ({response.status_code})\nResult: Failed to retrieve content")
            return False
    except requests.exceptions.RequestException as e:
        print(f"URL: {url}\nStatus: Error\nResult: Request failed: {e}")
        return False

# Test the URLs
url1 = "https://upload.wikimedia.org/wikipedia/commons/a/a3/Eq_it-na_pizza-margherita_sep2005_sml.jpg"
url2 = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/60/Sushi_platter.jpg/800px-Sushi_platter.jpg"

print("Checking URL 1:")
check_image_url(url1)

print("\nChecking URL 2:")
check_image_url(url2)
```

```python
import torchvision.transforms as transforms # Added import statement
from torch.utils.data import DataLoader # Ensure DataLoader is also imported if needed
import torchvision # Ensure torchvision is imported if needed

# --- 1. BETTER TRANSFORMS (Stop Squishing!) ---
stats = ((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))

train_transform = transforms.Compose([
    transforms.Resize(140),             # Resize smaller edge to 140
    transforms.CenterCrop(128),         # Crop the center 128x128
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2), # Stronger jitter
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])

test_transform = transforms.Compose([
    transforms.Resize(140),             # Resize smaller edge to 140
    transforms.CenterCrop(128),         # Crop the center 128x128
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])

# Reload the datasets with these new transforms
# This section will download the Food-101 dataset again if it was interrupted previously.
train_dataset = torchvision.datasets.Food101(root='./data', split='train',
                                             download=True, transform=train_transform)
test_dataset = torchvision.datasets.Food101(root='./data', split='test',
                                            download=True, transform=test_transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True,
                          num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False,
                         num_workers=2, pin_memory=True)

print("Data reloaded with Aspect Ratio Preservation.")
# Ensure this cell completes execution to define transforms and loaders.
```

```python
# --- 2. RELIABLE FINE-TUNING LOOP ---
import torch.optim as optim

# Reload model to start fresh (prevent bad weights from previous run)
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 101)
model = model.to(device)

# Unfreeze all layers
for param in model.parameters():
    param.requires_grad = True

# Optimizer & Scheduler
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-3)
scheduler = optim.lr_scheduler.OneCycleLR(optimizer, max_lr=1e-3,
                                          steps_per_epoch=len(train_loader),
                                          epochs=10) # 10 Epochs for better results
criterion = nn.CrossEntropyLoss()

print("Starting 10-Epoch Training...")

for epoch in range(10):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()

        # Gradient Clipping (Prevents the "Wrecking Ball" effect)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        optimizer.step()
        scheduler.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    print(f"Epoch [{epoch+1}/10] Train Acc: {100.*correct/total:.2f}% | Val Acc: {100.*val_correct/val_total:.2f}%")
```

### How to Load Your Saved Model

After a kernel disconnects, the Colab environment resets. However, your saved model in Google Drive persists. Here are the steps to load and reuse it:

1.  **Mount Google Drive**: You'll need to run `drive.mount('/content/drive')` again.
2.  **Re-define the Model Architecture**: When you save `model.state_dict()`, you're saving only the *weights* of the model, not the model's structure. You must recreate the same model architecture (e.g., `ResNet18` with 101 output classes) first.
3.  **Load the State Dictionary**: Load the saved `.pth` file and then load the `model_state_dict` into your freshly initialized model.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import os

# 1. Define paths explicitly in Drive
CHECKPOINT_PATH = "/content/drive/MyDrive/food101_checkpoint.pth"
BEST_MODEL_PATH = "/content/drive/MyDrive/food101_best_model.pth"

def train_model_smart(model, train_loader, test_loader, total_epochs=5, lr=0.001):

    # Initialize Optimizer & Loss
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)
    criterion = nn.CrossEntropyLoss()

    start_epoch = 0
    best_val_loss = float('inf')

    # --- RESUME LOGIC ---
    # Check if a checkpoint exists in Drive before starting
    if os.path.exists(CHECKPOINT_PATH):
        print(f"Checkpoint found at {CHECKPOINT_PATH}. Loading...")
        checkpoint = torch.load(CHECKPOINT_PATH)

        # 1. Load Model Weights
        model.load_state_dict(checkpoint['model_state_dict'])

        # 2. Load Optimizer State (Critical for momentum/Adam)
        # We need to ensure the loaded optimizer state matches the current optimizer's parameter groups.
        # If the model's trainable parameters have changed (e.g., from frozen to unfrozen backbone),
        # loading the old optimizer state might cause a mismatch.
        try:
            optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            print("  Optimizer state loaded successfully.")
        except ValueError as e:
            print(f"  WARNING: Failed to load optimizer state: {e}")
            print("  Optimizer state will be re-initialized from scratch. This is expected if trainable parameters have changed.")

        # 3. Load the epoch counter
        start_epoch = checkpoint['epoch'] + 1

        # 4. Load best loss so we don't overwrite a good model with a bad one
        best_val_loss = checkpoint.get('val_loss', float('inf'))

        print(f"Resuming training from epoch {start_epoch+1}/{total_epochs}")
    else:
        print("No checkpoint found. Starting training from scratch.")

    # If we already finished training, stop here
    if start_epoch >= total_epochs:
        print("Training already completed!")
        return model

    # --- TRAINING LOOP ---
    # Note: We loop from start_epoch to total_epochs
    for epoch in range(start_epoch, total_epochs):

        model.train()
        running_loss = 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # Validation phase
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for inputs, labels in test_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        # Calculate average losses
        avg_train_loss = running_loss / len(train_loader)
        avg_val_loss = val_loss / len(test_loader)

        print(f"Epoch [{epoch+1}/{total_epochs}] | "
              f"Train Loss: {avg_train_loss:.4f} | "
              f"Val Loss: {avg_val_loss:.4f}")

        # --- SAVE CHECKPOINT (Every Epoch) ---
        # This overwrites the file each time, acting as a 'save slot'
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(), # Save current optimizer state
            'val_loss': val_loss,
        }, CHECKPOINT_PATH)

        # --- SAVE BEST MODEL ---
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), BEST_MODEL_PATH)
            print("  New best model saved!")

    print("Training Complete.")
    return model
```
